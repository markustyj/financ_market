## Annual profit comparison

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
np.exp(10)
x = np.arange(0,12)
y0 = 50

def expo_func(x, power):
    return x*power

# y1 = np.power(x, 1.03) * y0
# y2 = np.power(x, 1.05) * y0
# y3 = np.power(x, 1.07) * y0
# y4 = np.power(x, 1.1) * y0
y1 = []
y2 = []
y3 = []
y4 = []

for i in range (len(x)):
    if i == 0:
        y1.append(y0)
        y2.append(y0)
        y3.append(y0)
        y4.append(y0)
    else:
        y1.append(expo_func(y1[i-1], 1.03) )
        y2.append(expo_func(y2[i-1], 1.05) )
        y3.append(expo_func(y3[i-1], 1.07) )
        y4.append(expo_func(y4[i-1], 1.1) )
        

        
        
fig = plt.figure()
ax = fig.add_subplot(111)        
        
ax.plot(x, y1)  
ax.plot(x, y2)
ax.plot(x, y3)
ax.plot(x, y4)
ax.grid()
ax.legend(["3%", "5%", "7%", "10%"] )
ax.set_xlabel("Year")
ax.set_ylabel("Full property / portofolio")


y = [y1,y2,y3,y4]
for yi in y:
    for i, y_element in enumerate(yi):
        ax.text(i, y_element+2, "%d" %y_element, ha="center")
```
